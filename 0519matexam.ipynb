```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
tdata = pd.read_csv("cases_add_Taiwan.csv",encoding='big5')
tdata
```

```python
Spots = tdata
Spots[['Date','Worldwide']].plot(kind = 'line', x = 'Date', y = 'Worldwide', title = "COVID-19 Daily Reports", legend = False)
plt.show()
```

```python
Spots = tdata
Spots[['Date','Taiwan']].plot(kind = 'line', x = 'Date', y = 'Taiwan', legend = False,x,y,color='red', lw=5, label="Taiwan")
Spots[['Date','China']].plot(kind = 'line', x = 'Date', y = 'China', legend = False,x,y,color='blue', lw=5, label="China")
Spots[['Date','Australia']].plot(kind = 'line', x = 'Date', y = 'Australia', legend = False,x,y,color='green', label="Australia")
plt.legend()
plt.title("COVID-19 Situation Reports")
plt.xlabel("Date")
plt.ylabel("cases")
plt.show()

```
